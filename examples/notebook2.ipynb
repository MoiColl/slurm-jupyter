# Some other analysis

```python
#%run standard.py
min_coverage = 3
```

Load libraries:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina', 'png')
sns.set()
sns.set_style('ticks')
```

Filter data:

```python
coverage = np.random.normal(5, size=2000)
coverage_filtered = coverage[coverage > min_coverage]
```

Some plot:

```python
x = np.linspace(0, 4, num=len(coverage_filtered))
y = coverage_filtered
plt.scatter(x, y, alpha=0.1)
plt.ylim(0, 8)
sns.despine()
```
